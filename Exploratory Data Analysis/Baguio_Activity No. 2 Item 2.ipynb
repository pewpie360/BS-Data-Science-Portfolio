# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('ggplot')
```

```python
# Create a n sample
n_samples = 100

age = np.random.randint(20, 65, size=n_samples)  # Random ages between 18 and 65
income = np.random.normal(50000, 15000, n_samples)  # Random incomes with a mean of 50,000 and standard deviation of 15,000

# Create a categorical data
gender = np.random.choice(['Male', 'Female'], size=n_samples)
education = np.random.choice(['High School', 'Bachelor', 'Master', 'Ph.D.'], size=n_samples)

df = pd.DataFrame({
    'Age': age,
    'Income': income,
    'Gender': gender,
    'Education': education
})
```

```python
print(df.head())
```

# Item Number 2 Problem

# Scatterplot

```python
plt.scatter(df['Age'], df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs. Income')
plt.show()
```

The scatter plot will show how 'Age' and 'Income' are distributed. If there's a visible trend or pattern, it might indicate a relationship between the two variables.

# Correlational

```python
correlation = df['Age'].corr(df['Income'])
print(f'Correlation between Age and Income: {correlation:.2f}')
```

The Pearson correlation coefficient ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation). A value close to 0 suggests no linear correlation. The result will tell you how strongly 'Age' and 'Income' are related.

# Regression

```python
from sklearn.linear_model import LinearRegression

X = df[['Age']]
y = df['Income']

regression_model = LinearRegression().fit(X, y)

# Get the coefficients
slope = regression_model.coef_[0]
intercept = regression_model.intercept_
print(f'Regression Equation: Income = {intercept:.2f} + {slope:.2f} * Age')
```

The regression equation will give you insights into how 'Income' changes with 'Age.' The slope indicates the rate of change.
