```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_1 = ["Samsung", "Huawei", "Xiaomi", "Apple", "Oppo", "Vivo", "Realme", "Others"]
#now to easily graph all data I add all quarters from 2018 to 2020 and divide each entry to 11 since there are 11 quarters overall
#except for realme since they start booming in year 2019 I just divide it to 7
y_1 = [20, 15, 9, 13, 8, 7, 3, 25]

fig1 = plt.figure(figsize = (5, 4))
axes = fig1.add_axes([0.1,0.1,0.9,0.9])

axes.set_title("Overall Global Smartphone Market Share (2018-2020)")
axes.set_xlabel("Smartphone brands")
axes.set_ylabel("Share rate")

axes.annotate("Other Smartphone brands continues\n to hold the top spot second to Samsung", xy = (5, 22), xytext = ("Huawei", 22))
                
axes = plt.plot(x_1,y_1)

axes
```

```python
fig2 = plt.figure(figsize = (5, 4))
axes_2 = fig2.add_axes([0.1,0.1,0.9,0.9])

axes_2.set_title("3 Years Market share percentage of samsung")

explode = [0] * 8
explode[0] = 0.2
axes_2 = plt.pie(y_1, explode = explode, labels = x_1, colors = ["blue", "grey", "skyblue", "grey", "skyblue", "grey", "skyblue", "grey",],
       autopct = "%1.0f%%", shadow = True, startangle = 140)
axes_2 = plt.legend(x_1, loc = "right", bbox_to_anchor = (1, 0, 0.5, 1))
```

```python
fig3 = plt.figure(figsize = (5, 4))
axes_3 = fig3.add_axes([0.1,0.1,0.9,0.9])

axes_3.set_title("Samsung only spot on the second of the highest\n market share rate for the past 3 years")
axes_3 = plt.barh(x_1, y_1, color = ["blue", "grey", "grey", "grey", "grey", "grey", "grey", "grey"])
```
