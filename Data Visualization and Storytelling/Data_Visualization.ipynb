# Data Visualization for titanic datasets 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic_df = pd.read_csv("titanic_dataset.csv")
titanic_df
```

# Bar Graph

```python
tds_df = pd.read_csv("titanic_dataset.csv")
tds_df = tds_df.sort_values(by="PassengerId")

tds_df1 = tds_df.values
x_1 = tds_df1[:,0]
y_1 = tds_df1[:,5]

fig_1 = plt.figure(figsize = (8, 6))
axes_1 = fig_1.add_axes([2,2,2,1])

axes_1.set_title("Age range or each Passenger")
axes_1.set_xlim([0,200])
axes_1.set_ylim([0,80])
axes_1.set_xlabel("PassengerId")
axes_1.set_ylabel("Age")
axes_1.bar(x_1, y_1)
```

# Violin Plot

```python
sns.violinplot(data = titanic_df, x = "Sex", y = "Age", hue = "Survived")
```
