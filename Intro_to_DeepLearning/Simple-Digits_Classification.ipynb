# Imports

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Use hand written dataset from keras lib

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#255 is the max gradient color so need to scale the data
X_train = X_train / 255 
X_test = X_test / 255
```

### Flattening the Dataset. 

We need to Transform 2D array of train into 1D because the first layer is a Flatten layer, which transforms the input images from a 2D array of pixels to a 1D array of numbers.

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
X_train_flattened.shape
```

### Creation of simple Neural Network 

* `Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments.`</br>

* `SparseCategoricalCrossentropy - Computes the crossentropy loss between the labels and predictions.`</br>

* `Accuracy - Calculates how often predictions equal labels.`</br>

* `Epochs - An epoch is when all the training data is used at once and is defined as the total number of iterations of all the training data in one cycle for training the machine learning model.`</br>

* `Relu - Applies the rectified linear unit activation function.`

```python
model = keras.Sequential([
    #Defining output and input neurons
    keras.layers.Dense(10, input_shape = (784, ), activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy', 
             metrics = ['accuracy'])

model.fit(X_train_flattened, y_train, epochs = 5)
```

### Model Evaluation 

```python
model.evaluate(X_train_flattened, y_train)
```

Model Evaluation shows that the model perfomance accuracy has 92% scoring. 

### Model Prediction 

```python
model.predict(X_test_flattened)
```

```python
plt.matshow(X_test[5])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[5] 
```

This shows 10 array predictions wich indicates the output neuron.

```python
# check for the max score in the array prediction and determine the actual prediction.
np.argmax(y_predicted[5])
```

```python
#convert predicted arrays into concrete class labels
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
#check for same output of truth data vs predicted data
y_test[:5]
```

### Check pred using confusion matrix

```python
conf_mat = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
conf_mat
```

```python
plt.figure(figsize = (20, 20))
sns.heatmap(conf_mat, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

### Add Hidden layer in Neural Network 

```python
model_1 = keras.Sequential([
    #Defining output and input neurons
    keras.layers.Dense(78, input_shape = (784, ), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model_1.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy', 
             metrics = ['accuracy'])

model_1.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model_1.evaluate(X_train_flattened, y_train)
```

```python
y_predicted = model_1.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
conf_mat1 = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)

plt.figure(figsize = (20, 20))
sns.heatmap(conf_mat1, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

### Keras feature for flattening the data layers. 

```python
model_2 = keras.Sequential([
    #Defining output and input neurons
    keras.layers.Flatten(input_shape = (28, 28)), #Flatten feature given the shape is 28 by 28
    keras.layers.Dense(78, activation = 'relu'), #Hidden layer with 78 ouput neurons
    keras.layers.Dense(10, activation = 'sigmoid') #Actual layer with 10 output neurons
])

model_2.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy', 
             metrics = ['accuracy'])

model_2.fit(X_train, y_train, epochs = 5)
```
