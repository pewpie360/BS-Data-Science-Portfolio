```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
social_med = ["LinkedIn", "Twitter", "Facebook"]
per = [10,70,90]

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])

axes.set_title("Audience spends their time on online")
axes = plt.barh(social_med, per, color = ["grey", "grey", "blue"])
```

```python
#saving figure to jpeg
fig.savefig("Plot_for_SocialMedia.jpeg")
```

```python
x_1 = ["internet meal ordering", "total restaurant sales", "other"]
y_1 = [20, 40, 40]

fig2 = plt.figure(figsize = (5, 4))
axes_2 = fig2.add_axes([0.1,0.1,0.9,0.9])

axes_2.set_title("Past 5 years of online food service")

explode = [0] * 3
explode[0] = 0.2
axes_2 = plt.pie(y_1, explode = explode, labels = x_1, colors = ["blue", "skyblue", "gray"],
       autopct = "%1.0f%%", shadow = True, startangle = 140)
axes_2 = plt.legend(x_1, loc = "right", bbox_to_anchor = (1, 0, 0.5, 1))
```
